```python
output = PolyChordOutput("/global/homes/s/sfbeltr/respaldo/piccaVersions/Vega/examples/C/out/", "qsoxlya")

outputB = PolyChordOutput("/global/homes/s/sfbeltr/respaldo/piccaVersions/Vega/examples/B/out/", "qsoxlya")



p_lya = [('ap', r'\alpha_{||}')]
p_lya += [('at', r'\alpha_\bot')]
p_lya += [('bias_eta_lya', r'b_{\eta Ly\alpha}')]
'''
p_lya += [('beta_lya', r'\beta_{Ly\alpha}')]
p_lya += [('bias_hcd', r'b_{HCD}')]
p_lya += [('beta_hcd', r'\beta_{HCD}')]
'''

posterior = get_post(output, p_lya)

posteriorB = get_post(outputB, p_lya)

sigma = 0.05

axis = 'at'

llimit = posterior.confidence(axis, sigma )
ulimit = posterior.confidence(axis, sigma, upper=True )

print( "lower = ", llimit  )  # 0.687   0.9545 
print( "upper = ", ulimit  )

llimitB = posteriorB.confidence(axis, sigma )
ulimitB = posteriorB.confidence(axis, sigma, upper=True )

print( "lower = ", llimitB  )  # 0.687   0.9545 
print( "upper = ", ulimitB  )

width  = ( ulimit - llimit )
widthB = ( ulimitB - llimitB )
ndiff = (width-widthB)/width

print( ndiff, " -> ", ndiff*100, "%" )
```
